결정계수(R제곱)은 뭘까? 회귀에서 테스트 결과가 얼마나 정확한지 나타내는 계수.

계산법: **R제곱= (타깃-예측)제곱의 합 나누기 (타깃-평균)제곱의 합**

결정계수는 예측이 타깃과 가까워질수록 1에 가까운 값이 된다.

# 규제

선형 회귀 모델의 경우, 특성에 곱해지는 계수(기울기)를 작게 만들어서 규제한다.

```python
#규제하기 전 특성 스케일을 정규화한다. 변환기 클래스인 StandardScaler 사용

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_poly)
train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)
```

```python
from sklearn.linear_model import Ridge #릿지: 계수를 제곱한 값을 기준으로 규제를 적용한다.
ridge = Ridge()
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))
```

```python
print(ridge.score(test_scaled, test_target))
```

릿지와 라쏘 모델을 사용할 때, alpha 매개변수를 통해 규제의 강도를 조절할 수 있다. 알파 값이 커서 규제의 강도가 세지면 과소적합될 가능성이 높다. 반대로 알파 값이 작아서 규제의 강도가 약해지면 과대적합될 가능성이 높다.
알파값은 하이퍼파라미터라서 사전에 우리가 지정해야 하는 하이퍼파라미터이다.

R제곱 값의 그래프를 그려서, 훈련 세트와 테스트 세트의 점수가 가장 가까운 지점이 최적의 alpha값이 된다.

```python
import matplotlib.pyplot as plt
train_score = []
test_score = []

#alpha 값을 10배씩 늘리면서.. 훈련세트와 테스트 세트의 점수를 리스트에 저장하기.
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
  ridge = Ridge(alpha=alpha)
  ridge.fit(train_scaled, train_target)

  train_score.append(ridge.score(train_scaled, train_target))
  test_score.append(ridge.score(test_scaled, test_target))
```

```python
plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.xlabel('alpha')
plt.ylabel( )
plt.show()
```


# 적절한 alpha 값: -1, 즉 10의 -1승인 0.1이다.
결정계수(R제곱)은 뭘까? 회귀에서 테스트 결과가 얼마나 정확한지 나타내는 계수.#

계산법: R제곱= (타깃-예측)제곱의 합 나누기 (타깃-평균)제곱의 합

결정계수는 예측이 타깃과 가까워질수록 1에 가까운 값이 된다.

```python
ridge = Ridge(alpha=0.1)
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))
print(ridge.score(test_scaled, test_target))
```

```python
from sklearn.linear_model import Lasso #라쏘는 계수의 절댓값을 기준으로 규제를 적용한다.
lasso = Lasso()
lasso.fit(train_scaled, train_target)
print(lasso.score(train_scaled, train_target))
```

```python
print(lasso.score(test_scaled, test_target))
```

```python
train_score=[]
test_score=[]
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
  lasso = Lasso(alpha=alpha, max_iter=10000)
  lasso.fit(train_scaled, train_target)

  train_score.append(lasso.score(train_scaled, train_target))
  test_score.append(lasso.score(test_scaled, test_target))
```

```python
plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()
```

```python
lasso = Lasso(alpha=10)
lasso.fit(train_scaled, train_target)
print(lasso.score(train_scaled, train_target))
print(lasso.score(test_scaled, test_target))
```
